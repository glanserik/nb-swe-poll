# Statistics from Swedish poll data

We begin by loading the necessary libraries and making som some settings.

```python
import pandas as pd
import numpy as np
import datetime as dt
import locale
locale.setlocale(locale.LC_ALL,'sv_SE.UTF-8') #This gives us Swedish dateformat, which is necessary for converting some
#strings to datetime

import matplotlib.pyplot as plt
#make inline plots
%matplotlib inline
import matplotlib.pylab as pylab
import matplotlib.lines as lines #to help in legend creation
pylab.rcParams['figure.figsize'] = 16, 12 #change the default size of the plot images
pylab.rcParams['figure.dpi'] = 300
```

We read the data from the csv-file, by giving pandas its url.

```python
url = 'https://raw.githubusercontent.com/MansMeg/SwedishPolls/master/Data/Polls.csv'
pollData = pd.read_csv(url)
pollData.head(5)
```

Next we want to calculate the mean value for each party, on a per month basis. After that we reset the index, this is because we don't have access to the data in the index field (which by default is the attribute we grouped by). We then create the new field published, and to fill it we convert the string in "PublYearMonth" to a datetime-object, and we also add 14 days so it will be "published" in the middle of each month. Lastly we set this as the new index.

```python
meanValues = pollData.groupby('PublYearMonth').mean()
meanValues = meanValues.reset_index()
meanValues['published'] = meanValues.PublYearMonth.apply(lambda date: dt.datetime.strptime(date,'%Y-%b') + dt.timedelta(days=14))
meanValues = meanValues.set_index('published')
```

We are also interested in the standard deviation, and we would like to see how it varies around the mean value. We therefore calculate the standard deviation and add/subtract it to the mean value, and then introduce the same "published" field as for the mean values.

```python

upperStd = pollData.groupby('PublYearMonth').std()+pollData.groupby('PublYearMonth').mean()
upperStd = upperStd.reset_index()
upperStd['published'] = upperStd.PublYearMonth.apply(lambda date: dt.datetime.strptime(date,'%Y-%b') + dt.timedelta(days=14))
upperStd = upperStd.set_index('published')

lowerStd = -pollData.groupby('PublYearMonth').std()+pollData.groupby('PublYearMonth').mean()
lowerStd = lowerStd.reset_index()
lowerStd['published'] = lowerStd.PublYearMonth.apply(lambda date: dt.datetime.strptime(date,'%Y-%b') + dt.timedelta(days=14))
lowerStd = lowerStd.set_index('published')

pollData['published'] = pollData.PublDate.fillna('1970-01-01').apply(lambda date: dt.datetime.strptime(date,'%Y-%m-%d'))
```

We now want to introduce a "published" field in the original data set. This should be based on the "PublDate" field. Right now we can't apply strptime() on this field, because some of the old data contains NaN, which is a float value (and strptime requires a string in the same format as the other strings in this field). To get around this we will replace all NaN in this column with the date string '1970-01-01'.

```python
pollData.PublDate = pollData.PublDate.fillna('1970-01-01')
```

Now we are able to use strptime to create the new field and set it as our index.

```python
pollData['published'] = pollData.PublDate.apply(lambda date: dt.datetime.strptime(date,'%Y-%m-%d'))
pollData = pollData.set_index('published')
```

After this we are ready to start plotting our data. In this example we want to see statistics for the Environment party after 2005.

```python
meanValues['2006-01-01':]['MP'].plot(linewidth=2,label='mean')
lowerStd['2006-01-01':]['MP'].plot(linestyle='--',color='slategray',linewidth=1,label='1 std around mean')
upperStd['2006-01-01':]['MP'].plot(linestyle='--',color='slategray',linewidth=1,label='1 std around mean')
pollData['2006-01-01':]['MP'].plot(style='g.',markersize=3)
meanline = lines.Line2D([], [], color='blue', linewidth=2, label='mean')
stdline = lines.Line2D([], [], color='slategray', linewidth=1,linestyle='--', label='1 std around mean')
markers = lines.Line2D(range(1), range(1), color="white", marker='.',markersize=3,markerfacecolor="green",label='individual polls')
plt.title('Poll results for MP from all avialable polls')
plt.xlabel('published')
plt.ylabel('%')
plt.legend(loc='best',handles=[meanline,stdline,markers])
```
